```python
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

rd_seed=123
```

```python
#make toy dataset no.1
from sklearn.datasets import make_blobs
X1, y1 = make_blobs(n_samples=1000, random_state=rd_seed)
print(X1.shape)
```

```python
#make toy dataset no.2
from sklearn.datasets import make_classification
X2, y2 = make_classification(n_samples=1000, n_features=2,n_redundant=0, n_clusters_per_class=1, random_state=rd_seed)
print(X2.shape)
```

```python
#make toy dataset no.3
from sklearn.datasets import make_circles
X3, y3 =  make_circles(n_samples=1000,noise=0.3,random_state=rd_seed)
print(X3.shape)
```

```python
#K-Means
from sklearn.cluster import KMeans

kmeans1 = KMeans(n_clusters=3).fit(X1)
y1_kmeans=kmeans1.predict(X1)
km_centroids1 = kmeans1.cluster_centers_

kmeans2 = KMeans(n_clusters=2).fit(X2)
y2_kmeans=kmeans2.predict(X2)
km_centroids2 = kmeans2.cluster_centers_

kmeans3 = KMeans(n_clusters=2).fit(X3)
y3_kmeans=kmeans3.predict(X3)
km_centroids3 = kmeans3.cluster_centers_

fig, axs = plt.subplots(1, 3)
axs[0].scatter(X1[:,0],X1[:,1],c=y1_kmeans)
axs[0].scatter(km_centroids1[:,0],km_centroids1[:,1], c='black', s=200, alpha=0.5)
axs[0].set_title('K Means - make_blobs')

axs[1].scatter(X2[:,0],X2[:,1],c=y2_kmeans)
axs[1].scatter(km_centroids2[:,0],km_centroids2[:,1], c='black', s=200, alpha=0.5)
axs[1].set_title('K Means - make_classification')

axs[2].scatter(X3[:,0],X3[:,1],c=y3_kmeans)
axs[2].scatter(km_centroids3[:,0],km_centroids3[:,1], c='black', s=200, alpha=0.5)
axs[2].set_title('K Means - make_circles')
fig.subplots_adjust(left=0.1, right=2)
```

```python
#Affinity Propagation
from sklearn.cluster import AffinityPropagation

affinity1 = AffinityPropagation(random_state=rd_seed, max_iter=2000).fit(X1)
y1_affinity = affinity1.predict(X1)
af_centroids1 = affinity1.cluster_centers_

affinity2 = AffinityPropagation(random_state=rd_seed, max_iter=2000).fit(X2)
y2_affinity = affinity2.predict(X2)
af_centroids2 = affinity2.cluster_centers_

affinity3 = AffinityPropagation(random_state=rd_seed, max_iter=2000).fit(X3)
y3_affinity = affinity3.predict(X3)
af_centroids3 = affinity3.cluster_centers_


fig, axs = plt.subplots(1, 3)
axs[0].scatter(X1[:,0],X1[:,1],c=y1_affinity)
#axs[0].scatter(af_centroids1[:,0],af_centroids1[:,1], c='black', s=200, alpha=0.5)
axs[0].set_title('AffinityPropagation - make_blobs')

axs[1].scatter(X2[:,0],X2[:,1],c=y2_affinity)
#axs[1].scatter(af_centroids2[:,0],af_centroids2[:,1], c='black', s=200, alpha=0.5)
axs[1].set_title('AffinityPropagation - make_classification')

axs[2].scatter(X3[:,0],X3[:,1],c=y3_affinity)
#axs[2].scatter(af_centroids3[:,0],af_centroids3[:,1], c='black', s=200, alpha=0.5)
axs[2].set_title('AffinityPropagation - make_circles')

fig.subplots_adjust(left=0.1, right=2)
```

```python
#SBSCAN
from sklearn.cluster import DBSCAN

dbscan1 = DBSCAN().fit(X1)
y1_dbscan=dbscan1.fit_predict(X1)

dbscan2 = DBSCAN().fit(X2)
y2_dbscan=dbscan2.fit_predict(X2)

dbscan3 = DBSCAN().fit(X3)
y3_dbscan=dbscan3.fit_predict(X3)

fig, axs = plt.subplots(1, 3)
axs[0].scatter(X1[:,0],X1[:,1],c=y1_dbscan)
axs[0].set_title('DBSCAN - make_blobs')

axs[1].scatter(X2[:,0],X2[:,1],c=y2_dbscan)
axs[1].set_title('DBSCAN - make_classification')

axs[2].scatter(X3[:,0],X3[:,1],c=y3_dbscan)
axs[2].set_title('DBSCAN - make_circles')
fig.subplots_adjust(left=0.1, right=2)
```

```python
#Gaussian mixture Model
from sklearn.mixture import GaussianMixture

gm1 = GaussianMixture(n_components=3, random_state=rd_seed).fit(X1)
y1_gm=gm1.predict(X1)

gm2 = GaussianMixture(n_components=2, random_state=rd_seed).fit(X2)
y2_gm=gm2.predict(X2)

gm3 = GaussianMixture(n_components=2, random_state=rd_seed).fit(X3)
y3_gm=gm3.predict(X3)

fig, axs = plt.subplots(1, 3)
axs[0].scatter(X1[:,0],X1[:,1],c=y1_gm)
axs[0].set_title('GaussianMixture - make_blobs')

axs[1].scatter(X2[:,0],X2[:,1],c=y2_gm)
axs[1].set_title('GaussianMixture - make_classification')

axs[2].scatter(X3[:,0],X3[:,1],c=y3_gm)
axs[2].set_title('GaussianMixture - make_circles')
fig.subplots_adjust(left=0.1, right=2)
```

```python
#BIRCH
from sklearn.cluster import Birch

birch1 = Birch(n_clusters=3).fit(X1)
y1_birch=birch1.predict(X1)

birch2 = Birch(n_clusters=2).fit(X2)
y2_birch=birch2.predict(X2)

birch3 = Birch(n_clusters=2).fit(X3)
y3_birch=birch3.predict(X3)

fig, axs = plt.subplots(1, 3)
axs[0].scatter(X1[:,0],X1[:,1],c=y1_birch)
axs[0].set_title('Birch - make_blobs')

axs[1].scatter(X2[:,0],X2[:,1],c=y2_birch)
axs[1].set_title('Birch - make_classification')

axs[2].scatter(X3[:,0],X3[:,1],c=y3_birch)
axs[2].set_title('Birch - make_circles')
fig.subplots_adjust(left=0.1, right=2)
```

```python
#Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering

agg1 = AgglomerativeClustering(n_clusters=3).fit(X1)
y1_agg=agg1.fit_predict(X1)

agg2 = AgglomerativeClustering(n_clusters=2).fit(X2)
y2_agg=agg2.fit_predict(X2)

agg3 = AgglomerativeClustering(n_clusters=2).fit(X3)
y3_agg=agg3.fit_predict(X3)

fig, axs = plt.subplots(1, 3)
axs[0].scatter(X1[:,0],X1[:,1],c=y1_agg)
axs[0].set_title('AgglomerativeClustering - make_blobs')

axs[1].scatter(X2[:,0],X2[:,1],c=y2_agg)
axs[1].set_title('AgglomerativeClustering - make_classification')

axs[2].scatter(X3[:,0],X3[:,1],c=y3_agg)
axs[2].set_title('AgglomerativeClustering - make_circles')
fig.subplots_adjust(left=0.1, right=2)
```

```python
#Mean Shift
from sklearn.cluster import MeanShift

meanshift1 = MeanShift().fit(X1)
y1_meanshift=meanshift1.predict(X1)

meanshift2 = MeanShift().fit(X2)
y2_meanshift=meanshift2.predict(X2)

meanshift3 = MeanShift().fit(X3)
y3_meanshift=meanshift3.predict(X3)

fig, axs = plt.subplots(1, 3)
axs[0].scatter(X1[:,0],X1[:,1],c=y1_meanshift)
axs[0].set_title('MeanShift - make_blobs')

axs[1].scatter(X2[:,0],X2[:,1],c=y2_meanshift)
axs[1].set_title('MeanShift - make_classification')

axs[2].scatter(X3[:,0],X3[:,1],c=y3_meanshift)
axs[2].set_title('MeanShift - make_circles')
fig.subplots_adjust(left=0.1, right=2)
```
